```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
table_nilmtk=pd.DataFrame(index=np.arange(1,796),columns=['Index','Issue Name','# of Comments','Closed','Participants','Category'])
```

```python
ind=0
for j in tqdm(range(1,30)):    
    link='https://github.com/nilmtk/nilmtk/issues?page='+str(j)+'&q=is%3Aissue+sort%3Acreated-asc'
    #print(link)
    #break
    page = requests.get(link)
    soup = BeautifulSoup(page.text, 'html.parser')
    for i in range(1,26):
        issue=soup.find("div", {"id": "issue_"+str(ind+1)})
        result1 = re.search('href="/nilmtk/nilmtk/issues/(.*)" id="issue', str(issue)) #number
        result2 = re.search('link">(.*)</a>', str(issue)) #name/
        result3 = re.search('class="text-small text-bold">(.*)</span>', str(issue)) # Comments    
        result4 = re.search('aria-label="(.*) issue"', str(issue)) #Closed/Open
        result5 = re.findall('color: #......" title="(.*)">', str(issue)) #Category
        if result1==None:
            number="None"
        else:
            number=result1.group(1)
        if result2==None:
            name="None"
        else:
            name=result2.group(1)
        if result3==None:
            comment=0
        else:
            comment=result3.group(1)
        if result4==None:
            closed="None"
        else:
            closed=result4.group(1)    
        if result5==None:
            category="None"
        else:
            category=result5
        
        link_new='https://github.com/nilmtk/nilmtk/issues/'+str(number)
        page_new = requests.get(link_new)
        soup_new = BeautifulSoup(page_new.text, 'html.parser')
        issue_new=soup_new.find("div",id="partial-users-participants")
        result6 = re.search('(.*) participant', str(issue_new)) #participants
        if result6==None:
            participants=1
        else:   
            participants=result6.group(1)[6]
        
        #print(number,name,comment,closed,category,participants)
        table_nilmtk.iloc[ind]['Index']=number 
        table_nilmtk.iloc[ind]['Issue Name']=name 
        table_nilmtk.iloc[ind]['# of Comments']=comment 
        table_nilmtk.iloc[ind]['Closed']=closed 
        table_nilmtk.iloc[ind]['Participants']=participants
        table_nilmtk.iloc[ind]['Category']= category
        ind=ind+1
    print(ind)    
    if ind==100:
        break
```
