# Gradient Descent

```python
x=3
# Parameters required for Gradient Descent
alpha = 0.1   #learning rate
np.random.seed(10)  
def gradient_descent(x,alpha):
    cost_list=[]
    theta_list=[]
    prediction_list=[]
    cost_list.append(1e2)#large inital cost=10^2
    run=True
    i=0
    #iterating gradient descent
    while run:
        cost=x #cost=x
        cost_list.append(cost)
        x=x-(alpha*2*x) #x=x- alpha * 2*x
        if cost_list[i]-cost_list[i+1]< 1e-9:#checking if the change in cost function is less than 10^(-9)
            run=False
        i=i+1
    cost_list.pop(0)#remove intital cost 
    return prediction_list, cost_list, theta_list
```

```python
prediction_list, cost_list, theta_list = gradient_descent(x,alpha)
```

```python
#Final Cost
cost_list[-1]
```

# Cost Function

```python
#plot of decreasing cost
plt.plot(cost_list)
plt.show()
```

# Code for Part 3

```python
alpha = 0.1   #learning rate
np.random.seed(10) 
x1=3
x2=3
def gradient_descent(x1,x2,alpha):
    cost_list=[]
    theta_list=[]
    prediction_list=[]
    cost_list.append(1e2)#large inital cost=10^2
    run=True
    i=0
    #iterating gradient descent
    while run:
        cost=x1**2+x2**2 #cost=x1^2+x2^2
        cost_list.append(cost)
        x1=x1-(alpha*2*x1) #x1=x1- alpha * 2 * x1
        x2=x2-(alpha*2*x2) #x2=x2- alpha * 2 * x2
        if cost_list[i]-cost_list[i+1]< 1e-9:#checking if the change in cost function is less than 10^(-9)
            run=False
        i=i+1
    cost_list.pop(0)#remove intital cost 
    return prediction_list, cost_list, theta_list
prediction_list, cost_list, theta_list = gradient_descent(x1,x2,alpha)
```

```python
#Plotting of f(x1,x2)=x1^2+x2^2
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(1,1,1,projection='3d')
a0 = np.linspace(-5,5,100)
a1 = np.linspace(-5,5,100)
aa0, aa1 = np.meshgrid(a0, a1)
ax.plot_surface(aa0, aa1, aa0*aa0+aa1*aa1)
```
